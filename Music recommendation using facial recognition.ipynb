```python
import cv2
from deepface import DeepFace
import tkinter as tk
from tkinter import *
from playsound import playsound
from pygame import mixer

def stop_music():
    mixer.music.stop()

def display_camera():
    global gender
    global dominant_emotion
    FaceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        raise IOError('Cannot open webcam')
    
    while True:
        ret, frame = cap.read()
        results = DeepFace.analyze(frame, actions=['emotion', 'gender'], enforce_detection=False)
        if isinstance(results, list):
            # If multiple faces detected, process each face separately
            for result in results:
                gender_result = result['gender']
                dominant_emotion = result['dominant_emotion']
                # Add your face detection and drawing logic here
        else:
            # If only one face detected, process the result
            gender_result = results['gender']
            dominant_emotion = results['dominant_emotion']
            # Add your face detection and drawing logic here
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = FaceCascade.detectMultiScale(gray, 1.1, 4)
        for x, y, w, h in faces:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(frame, dominant_emotion, (50, 50), font, 1, (0, 0, 255), 2, cv2.LINE_4)
        cv2.imshow("Webcam", frame)
        if cv2.waitKey(2) & 0xFF == ord('q'):
            break
  
    cap.release()
    cv2.destroyAllWindows()


def play_music():
    global gender
    global dominant_emotion
    if var.get() == 1:
        if gender == 'Man':
            sound_file = 'Male.mp3'
        elif gender == 'Woman':
            sound_file = 'Female.mp3' 
        mixer.init() 
        mixer.music.load(sound_file) 
        mixer.music.play()
    elif var.get() == 2:
        if dominant_emotion == 'happy':
            sound_file = 'Happy.mp3'
        elif dominant_emotion == 'angry':
            sound_file = 'angry.mp3'
        elif dominant_emotion == 'sad':
            sound_file = 'sad.mp3'
        elif dominant_emotion == 'neutral':
            sound_file = 'neutral.mp3'
        mixer.init()
        mixer.music.load(sound_file)
        mixer.music.play()

root = tk.Tk()
root.title("Emotion Detection and Song Recommendation")
root.geometry("500x500")

label_title = tk.Label(root, text='Emotion Detection and Song Recommendation', font=("Arial", 15))
label_title.pack(pady=10)

label1 = tk.Label(root, text='Name:')
label1.pack()
entry1 = tk.Entry(root)
entry1.pack()

button1 = tk.Button(root, text='Capture Emotion', command=display_camera, bg='orange')
button1.pack(pady=10)

var = IntVar()
label_radio = tk.Label(root, text='Play Music By:')
label_radio.pack()
R1 = Radiobutton(root, text="Gender", variable=var, value=1, bg='light blue', command=play_music)
R1.pack()
R2 = Radiobutton(root, text="Emotion", variable=var, value=2, bg='light blue', command=play_music)
R2.pack()

root.mainloop()
```
